# Community Area Populations

<h3>Question 2a: Merge the following data sets: the census tract population data, and the Community Area tracts mapping.</h3><p>Join these together using the fact that the last six digits of the tract ID in the mapping data correspond to the first six digits of the block ID. However, the data portal has a bug: if the block starts with a zero, that digit is missing!</p>

```python
ctracts = pd.read_csv("https://raw.githubusercontent.com/yarikan/assignments/master/community_tracts.csv?token=AQAvwLqvANWK719Ri_c6kJQ14acDB9-kks5WuhupwA%3D%3D")
# ctracts.head(2)  # To merge, I need the last six digits of the tract id
```

```python
# make sure tract IDs are of a type that I can slice
print(type(ctracts['tract_id'][0]))
```

```python
# integers are not slicable, convert to string
ctracts['tract_id'] = ctracts['tract_id'].apply(str)
```

```python
# define function to edit tract IDs and keep only the last six digits
def extract_ctracts_id(var):
    return var[5:]
```

```python
ctracts['tract_id'] = ctracts['tract_id'].apply(extract_ctracts_id)
# ctracts['tract_id']
```

```python
cblocks = pd.read_csv("https://raw.githubusercontent.com/yarikan/assignments/master/Population_by_2010_Census_Block.csv?token=AQAvwCF1xGHA5EWmhErsOofYDeDGjlC4ks5Wuhv3wA%3D%3D")
# cblocks.head(2)  # block id has either 9 or 10 digits. To merge, I need the first six
```

```python
print(type(cblocks['CENSUS BLOCK'][0]))
```

```python
cblocks['CENSUS BLOCK'] = cblocks['CENSUS BLOCK'].apply(str)
```

```python
# write function to assure that all census block IDs have 10 digits, before proceeding to slice
def complete_to_10digits(var):
    if len(var) == 9:
        return '0' + var
    if len(var) == 10:
        return var
```

```python
cblocks['CENSUS BLOCK'] = cblocks['CENSUS BLOCK'].apply(complete_to_10digits)
# cblocks['CENSUS BLOCK']
```

```python
# define function to slice census block IDs and keep only first 6 digits
def extract_cblock_id(var):
    return var[:6]
```

```python
cblocks['CENSUS BLOCK'] = cblocks['CENSUS BLOCK'].apply(extract_cblock_id)
# cblocks['CENSUS BLOCK']
```

```python
blocks_and_tracts = cblocks.merge(ctracts, left_on=['CENSUS BLOCK'], right_on=['tract_id'], how='inner')
blocks_and_tracts.sort(' community_id')
```

<p>The data frame above (blocks_and_tracts) is the answer to Question 2a.</p>
